```
# Load pretrained models
from notebook_utils.synthesize import *
import IPython.display as ipd
init_hparams('notebook_utils/pretrained_hparams.py')
tts_model = get_forward_model('pretrained/forward_400K.pyt')
voc_melgan = get_melgan_model() 
voc_wavernn = get_wavernn_model('pretrained/wave_575K.pyt')
```

##Synthesize

```
# Synthesize with melgan (alpha=1.0)
input_text = 'I'
wav = synthesize(input_text, tts_model, voc_melgan, alpha=1)
ipd.Audio(wav, rate=hp.sample_rate)
```

```
# Synthesize with wavernn (alpha=1.0)
input_text = 'President Trump met with other leaders at the group of 20 conference.'
wav = synthesize(input_text, tts_model, voc_wavernn, alpha=1)
ipd.Audio(wav, rate=hp.sample_rate)
```

```
# Synthesize with faster speed (alpha=1.2)
input_text = 'In the middle of difficulty lies opportunity.'
wav = synthesize(input_text, tts_model, voc_wavernn, alpha=1.2)
ipd.Audio(wav, rate=hp.sample_rate)
```

```
# Synthesize with slower speed (alpha=0.8)
input_text = 'In the middle of difficulty lies opportunity.'
wav = synthesize(input_text, tts_model, voc_wavernn, alpha=0.7)
ipd.Audio(wav, rate=hp.sample_rate)
```
